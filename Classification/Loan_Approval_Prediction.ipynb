```python
error_cols = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']
```

```python
for c in error_cols:
    q1 = dfloan[c].quantile(0.25)
    q3 = dfloan[c].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - (1.5 * iqr )
    upper = q3 + (1.5 * iqr )
    dfloan[c] = dfloan[c].map(lambda x :  lower if x < lower else  upper  if x > upper else x)
```

```python
sns.boxplot(dfloan)
plt.xticks(rotation = 90)
plt.show()
```
