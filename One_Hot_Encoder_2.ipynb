### Ordinal Encoder

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
ord = OrdinalEncoder()
```

```python
ord
```

```python
ord.fit(dfmovie[["education"]])
```

```python
ord.transform(dfmovie[["education"]])
```

```python
ary_edu = ord.transform(dfmovie[["education"]])
```

```python
ary_edu
```

```python
ord.categories_
```

```python
dfmovie.head()
```

```python
dfmovie["education"] = ary_edu
```

### Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
```
