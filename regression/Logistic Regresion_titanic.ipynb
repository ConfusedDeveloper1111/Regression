```python
sclist = []
for i in range(1, 1000):
    Xtrain, Xtest, ytrain, ytest = train_test_split(X_ros, y_ros, test_size= 0.20, random_state= i)
    lrmodel = LogisticRegression()
    lrmodel.fit(Xtrain, ytrain)
    sc = lrmodel.score(Xtest, ytest)
    sclist.append(sc)

m = max(sclist)
rs = sclist.index(m)
print(rs, "==>", m)
```
