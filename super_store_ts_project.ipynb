### SARIMAX

```python
p = list(range(2))
p
```

```python
d = [0,1]
d
```

```python
q = p.copy()
p
```

```python
p,d,q
```

```python
import itertools
```

```python
itertools.product(p,d,q)
```

```python
list(itertools.product(p,d,q))
```

```python
pdq = list(itertools.product(p,d,q))
pdq
```

```python
import time
for i in pdq:
    print(i)
    time.sleep(1)
```

```python
(0,0,1)+(12,)
```

```python
for i in pdq:
    print(i + (12,1))
```

```python
[i + (12,) for i in pdq]
```

```python
spdq = [i + (12,) for i in pdq]
spdq
```

```python
for p1 in pdq:
    for p2 in spdq:
        print(p1, 'x', p2)
```

```python
import statsmodels.api as sm
```

```python
for p1 in pdq:
    for p2 in spdq:
        ar =sm.tsa.SARIMAX(y, orders= p1, seasonal_order= p2, enforce_stationarity= False, enforce_invertibility= False)
        result = ar.fit()
        print(result.aic)
```

```python
for p1 in pdq:
    for p2 in spdq:
        ar =sm.tsa.SARIMAX(y, orders= p1, seasonal_order= p2, enforce_stationarity= False, enforce_invertibility= False)
        result = ar.fit()
        print(p1,'x', p2,' ==> ', result.aic)
```

```python
aic_value = 1000
for p1 in pdq:
    for p2 in spdq:
        try:
           ar =sm.tsa.SARIMAX(y, orders= p1, seasonal_order= p2, enforce_stationarity= False, enforce_invertibility= False)
           result = ar.fit()
           if result.aic < aic_value:
              aic_value = result.aic
              order1 = p1
              order2 = p2
        except:
            print("triggered error")
print("aic_value", aic_value)
print("order1", order1)
print("order2", order2)
```

```python
armodel = sm.tsa.SARIMAX(y, order= (0,1,1), seasonal_order = (0,1,1,12),enforce_invertibilty = False)
```

```python
result = armodel.fit()
```
